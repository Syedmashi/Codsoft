```python
data1=data.drop(['species'],axis=1)
data1.head()
```

```python
data2=data['species']
data2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(data1,data2, test_size = 0.33)
```

```python
y_test.head()
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
model.predict(X_test)
```

```python
accuracylog = model.score(X_test, y_test)
print(f"Accuracy: {accuracylog:.2f}")
```

```python
modelsvc = SVC()
modelsvc.fit(X_train, y_train)
modelsvc.predict(X_test)
```

```python
accuracy_svc = model.score(X_test, y_test)
print(f"Accuracy: {accuracy_svc:.2f}")
```

```python
modelknn = KNeighborsClassifier()
modelknn.fit(X_train, y_train)
modelknn.predict(X_test)
```

```python
accuracy_knn = modelknn.score(X_train, y_train)
print(f"Accuracy: {accuracy_knn:.2f}")
```

```python
modelgauss = GaussianNB()
modelgauss.fit(X_train, y_train)
modelgauss.predict(X_test)
```

```python
accuracy_gauss = modelgauss.score(X_train, y_train)
print(f"Accuracy: {accuracy_gauss:.2f}")
```

```python
model_decision_tree = DecisionTreeClassifier()
model_decision_tree.fit(X_train, y_train)
model_decision_tree.predict(X_test)
```

```python
acc_decision_tree =model_decision_tree.score(X_train, y_train)
print(f"Accuracy: {acc_decision_tree:.2f}")
```

```python
model_random_forest = RandomForestClassifier(n_estimators=100)
model_random_forest.fit(X_train, y_train)
model_random_forest.predict(X_test)
```

```python
acc_random_forest =model_random_forest.score(X_train, y_train)
print(f"Accuracy: {acc_random_forest:.2f}")
```

```python
models = pd.DataFrame({
    'Model': ['Logistic Regression','SVC','K Neighbors Classifier','Gaussian NB','Decision Tree Classifier',
              'Random Forest Classifier'],
    'Score': [accuracylog,accuracy_svc,accuracy_knn,accuracy_gauss,acc_decision_tree,acc_random_forest]})
models.sort_values(by='Score', ascending=False)
```
